# K-Means Clustering

# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing Datasets

```python
datasets = pd.read_csv('Mall_Customers.csv')
X = datasets.iloc[:,[3,4]].values
```

# Using the elbow methord to find optimal number of clusters

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans  = KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(X) 
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title("The Elbow methord")
plt.xlabel("k values")
plt.ylabel("Wcss values")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5,init='k-means++',random_state=42)
Y_kmeans = kmeans.fit_predict(X)
```

```python
print(Y_kmeans)
```

# visualizing the clusters

```python
plt.scatter(X[Y_kmeans==0,0],X[Y_kmeans==0,1], s=100/2,c='violet',label='cluster1')
plt.scatter(X[Y_kmeans==1,0],X[Y_kmeans==1,1], s=100/2,c='indigo',label='cluster2')
plt.scatter(X[Y_kmeans==2,0],X[Y_kmeans==2,1], s=100/2,c='blue',label='cluster3')
plt.scatter(X[Y_kmeans==3,0],X[Y_kmeans==3,1], s=100/2,c='green',label='cluster4')
plt.scatter(X[Y_kmeans==4,0],X[Y_kmeans==4,1], s=100/2,c='yellow',label='cluster5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1] ,s=300/2,c='orange',label='centre')
plt.title("Cluster of customers")
plt.xlabel("Annual income")
plt.xlabel("Spending Score")
plt.legend()
plt.show()
```
